## 2.1. Graphe 1

Réaliser un graphique avec deux axes. L'axe des abscisses va de 0 à 50 et s'intitule "Axe des x". L'axe des ordonnées s'intitule "Axe des y".
Tracer la droite d'équation : x = 3y. Avec x allant de 0 à 50.
Le graphe possède le titre suivant "Tracer une droite" et la droite est de couleur rouge.

```python
# Create the vectors X and Y
x = np.array(range(50))
y = x / 3

# Create the plot and choose a color
plt.plot(x, y, color='red', label='y = x/3')

# Add a title
plt.title("Tracer une droite")

# Add X and y Label
plt.xlabel('x axis')
plt.ylabel('y axis')

# Set axis range
plt.xlim([0, 50])
plt.ylim([0, 20])

# Add a grid
plt.grid(alpha=.2,linestyle='--')

# Add a Legend which is added from the 'label'
plt.legend()

# Show the plot
#plt.plot(x, y)
plt.show()

```

## 2.7. Graphe 7

Nous souhaitons un graphe qui donne l'évolution du CA pour chaque catégorie (en fonction du temps).

```python
df_transactions
```

```python
df_transactions = pd.merge(df_transactions,df_products,on='id_prod',how='left')
df_transactions = df_transactions.reindex(columns=['id_prod', 'price', 'categ', 'date', 'session_id','client_id'])
```

```python
df_transactions["month"] = df_transactions["date"].str[5:7]
df_transactions["year"] = df_transactions["date"].str[:4]
df_transactions["month_year"] = df_transactions["month"] + '-' + df_transactions["year"]
print(df_transactions["month_year"])
```

```python
df_transactions_v2 = df_transactions[['month_year', 'categ', 'price']]
df_transactions_v2 = df_transactions_v2.dropna(subset =['categ'])
df_transactions_v2["Revenue"] = df_transactions_v2.groupby(['month_year',"month_year"])['price'].transform('sum')
df_transactions_v2 = df_transactions_v2.drop(columns=["price"])
df_transactions_v2 = df_transactions_v2.drop_duplicates()
df_transactions_v2
```

```python
df_transactions_categ_0 = df_transactions_v2.loc[df_transactions_v2['categ'] == 0].drop(columns=["categ"])
df_transactions_categ_0['month_year'] = pd.to_datetime(df_transactions_categ_0["month_year"],format='%m-%Y',errors='coerce')
df_transactions_categ_0 = df_transactions_categ_0.sort_values('month_year')
df_transactions_categ_1 = df_transactions_v2.loc[df_transactions_v2['categ'] == 1].drop(columns=["categ"])
df_transactions_categ_1['month_year'] = pd.to_datetime(df_transactions_categ_1["month_year"],format='%m-%Y',errors='coerce')
df_transactions_categ_1 = df_transactions_categ_1.sort_values('month_year')
df_transactions_categ_2 = df_transactions_v2.loc[df_transactions_v2['categ'] == 2].drop(columns=["categ"])
df_transactions_categ_2['month_year'] = pd.to_datetime(df_transactions_categ_2["month_year"],format='%m-%Y',errors='coerce')
df_transactions_categ_2 = df_transactions_categ_2.sort_values('month_year')
df_transactions_categ_2
```

```python
fig = plt.figure(figsize =(10, 7))
plt.plot(df_transactions_categ_0["month_year"], df_transactions_categ_0["Revenue"], color='#d0f4de', label='Cat:0')
plt.plot(df_transactions_categ_1["month_year"], df_transactions_categ_1["Revenue"], color='#e4c1f9', label='Cat:1')
plt.plot(df_transactions_categ_2["month_year"], df_transactions_categ_2["Revenue"], color='#a9def9', label='Cat:2')

plt.title("Evolution du CA par catégorie", fontweight="bold")
plt.legend()
ax = plt.subplot()


plt.xlabel('mois')
plt.ylabel('profit (en $)')


plt.grid(alpha=.4,linestyle='--')

months = ['03/2021', '04/2021', '05/2021', '06/2021', '07/2021', '08/2021', '09/2021', '10/2021', '11/2021', '12/2021', '01/2022', '02/2022']
plt.xticks(df_transactions_categ_0["month_year"], months, rotation = 45)

plt.show()
```
